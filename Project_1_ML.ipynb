```python
data_copy = data.copy(deep=True)
```

```python
data_copy.columns
```

```python
data_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']]= data_copy[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI']].replace(0,np.nan)
```

```python
data_copy.isnull().sum()
```

```python
data_copy.head()
```

```python
data_copy['Insulin'] = data_copy['Glucose'].replace(np.nan,data_copy['Insulin'].mean())
```

In the main copy Data columns like Glucose, BloodPressure,SkinThickness,	Insulin,BMI,null are values are hidden behind 0 values, we can predict the right values those being 0, taking copy of the data we replaced 0 with NaN and later detecting those NaN values replacing them with mean of the columns

```python
data['Glucose'] = data['Glucose'].replace(0,data['Glucose'].mean())
data['BloodPressure'] =  data['BloodPressure'].replace(0,data['BloodPressure'].mean())
data['SkinThickness'] =  data['SkinThickness'].replace(0,data['SkinThickness'].mean())
data['Insulin'] =  data['Insulin'].replace(0,data['Insulin'].mean())
data['BMI'] =  data['BMI'].replace(0,data['BMI'].mean())
```

```python
data.head()
```

```python
columns_with_zeros = data_copy.columns[(data_copy == 0).any()]

# Calculate the sum of columns with zeros
sum_of_columns = data_copy[columns_with_zeros].sum()

print("Columns with 0 values:", columns_with_zeros)
print("Sum of columns with 0 values:", sum_of_columns)
```

```python
#Store feature Matrix in X and Response (Target) in vector y
```

```python
x = data.drop('Outcome',axis=1)
y = data['Outcome']
```

```python
x
```

```python
y
```

```python
#Splitting the dataset into the trainig set and test set
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test= train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#Scikit Learn Pipeline
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.pipeline import Pipeline
```

```python
pipeline_lr = Pipeline([('scalar1',StandardScaler()),('lr_classifier',LogisticRegression())])

pipeline_knn = Pipeline([('scalar2',StandardScaler()),('knn_classifier',KNeighborsClassifier())])

pipeline_svc = Pipeline([('scalar3',StandardScaler()),('svc_classifier',SVC())])

pipeline_dt = Pipeline([('dt_classifier',DecisionTreeClassifier())])

pipeline_rf = Pipeline([('rf_classifier',RandomForestClassifier(max_depth=3))])

pipeline_gbc = Pipeline([('gbc_classifier',GradientBoostingClassifier())])
```

```python
Pipelines = [pipeline_lr,pipeline_knn,pipeline_svc,pipeline_dt,pipeline_rf,pipeline_gbc]
```

```python
Pipelines
```

```python
for pipe in Pipelines:
  pipe.fit(x_train,y_train)
```

```python
pipe_dict = {0:'LR',
             1:'KNN',
             2:'SVC',
             3:'DT',
             4:'RF',
             5:'GBC'
             }
```

```python
pipe_dict
```

```python
for i,  model in enumerate(Pipelines):
  print("{} Test Accuracy:{}".format(pipe_dict[i],model.score(x_test,y_test)*100))
```
